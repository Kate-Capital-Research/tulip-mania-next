```python
%load_ext autoreload
%autoreload 2

from tulip.data.fred import get_series

import plotly.graph_objects as go

from pandas.tseries.offsets import MonthBegin, MonthEnd
from tulip.themes import blog_theme as THEME
from tulip.plots import add_plot_theme
import pyfredapi as pf
```

```python
# Pull in data
core_pce_level = get_series(series_id="PCEPILFE")
# This natively pulls in what is called a DataFrame
core_pce_level.head(5)
```

```python
# We transform the dataframe into a series (we only want value)
core_pce_level = core_pce_level.set_index('date')['value']
core_pce_level.tail(5)
```

```python
core_pce_change = core_pce_level.pct_change(6) * 2
core_pce_change = core_pce_change.dropna()
core_pce_change
```

```python
# if in doubt...
# core_pce_level.to_clipboard()
```

```python
# Plotly
fig = go.Figure()

# median line
fig.add_trace(go.Scatter(
    x=core_pce_change.index,
    y=core_pce_change,
    mode='lines',
    name='US CORE PCE (6m change, ann., SA)',
    # line=dict(color=BC.LINE)
))

fig = add_plot_theme(
    fig,
    title='Sticky inflation is falling rapidly', 
    subtitle='US CORE PCE', 
    source='FRED, Still Press',
    x_axis_label='', 
    y_axis_label='',
    extend_months = 12,
    theme='fund'
)

fig
```
